```python
import matplotlib.pyplot as plt

year = [2005, 2010, 2015, 2020]
single_person = [20.0, 25.0, 30.0, 35.0]

plt.figure(figsize=(12, 8))

plt.subplot(221)
plt.bar(year, single_person)
plt.title('Bar')

plt.subplot(222)
plt.plot(year, single_person, marker='o')
plt.title('Line 1')

plt.subplot(223)
plt.plot(year, single_person, color='g', marker='x')
plt.title('Line 2')

plt.subplot(224)
plt.scatter(year, single_person)
plt.title('Scatter')


plt.show()
```

```python
import matplotlib.pyplot as plt
import random

dice1 = plt.imread('dice1.png')
dice2 = plt.imread('dice2.png')
dice3 = plt.imread('dice3.png')
dice4 = plt.imread('dice4.png')
dice5 = plt.imread('dice5.png')
dice6 = plt.imread('dice6.png')
dices = [dice1, dice2, dice3, dice4, dice5, dice6]

num_1 = random.randint(1, 6)
num_2 = random.randint(1, 6)

plt.figure(figsize=(5,3))

plt.subplot(1,2,1)
# x y축 없애줌
plt.xticks([])
plt.yticks([])
plt.xlabel(num_1)
plt.imshow(dices[num_1-1])
#plt.axis('off')

plt.subplot(1,2,2)
plt.xticks([])
plt.yticks([])
plt.xlabel(num_2)
plt.imshow(dices[num_2-1])
#plt.axis('off')
```
