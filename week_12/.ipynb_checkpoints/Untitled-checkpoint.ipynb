# 문제1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('학급당_학생수.csv', encoding = 'EUC-KR')
display(df)
```

```python
ind  = np.arange(len(df))
from matplotlib import rc
rc("font", family='Malgun Gothic')
rc("font", size=10)
plt.figure(figsize=(5,3))
plt.plot(ind, df['서울특별시'], label = 'seoul')
plt.plot(ind, df['제주특별자치도'], label = 'jeju')
plt.xticks(ind, df['시점'])
plt.ylim(15, 30)
plt.legend()
plt.xlabel('연도')
plt.xlabel('학급당 학생 수')
plt.xticks(rotation = 'vertical')
plt.show()
```

# 문제2

```python
from matplotlib import font_manager
f_path='c:/Windows/Fonts/malgun.ttf'
font9=font_manager.FontProperties(fname=f_path, size=9)
font11=font_manager.FontProperties(fname=f_path, size=11)
import numpy as np
import pandas as pd

df1 = pd.read_csv('지역별면적.csv', encoding = 'EUC-KR')
df2 = pd.read_csv('지역별인구수.csv', encoding = 'EUC-KR')

df = pd.merge(df1, df2, on='지역')

df['인구밀도'] = df['총인구수']/df['면적']
display(df[df['인구밀도']>=2000])
ind = np.arange(len(df))
plt.figure(figsize=(5,3))
plt.bar(ind, df['인구밀도'])

plt.xlabel('지역', fontproperties=font11)
plt.ylabel('인구밀도', fontproperties=font11)
plt.xticks(ind, df['지역'], rotation='vertical', fontproperties=font9)
plt.show()
```

# 문제 3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('한국에너지공단_자동차 표시연비 정보.csv')
display(df.head())

df = df[df['배기량']!=0]

df1 = df[df['연료']=='휘발유']
df2 = df[df['연료']=='경유']

plt.figure(figsize=(5,3))
plt.scatter(df1['배기량'], df1['CO2배출량(g_km)'], s=3, alpha=0.5, label='Gasoline')
plt.scatter(df2['배기량'], df2['CO2배출량(g_km)'], s=3, alpha=0.5, label='Diesel')
plt.legend()
plt.xlabel('displacement')
plt.ylabel('CO2 emissions')
plt.show()

cor = df1['배기량'].corr(df1['CO2배출량(g_km)'])
print(cor)

cor = df2['배기량'].corr(df2['CO2배출량(g_km)'])
print(cor)
```
