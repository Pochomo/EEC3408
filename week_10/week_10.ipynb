# 문제 \# 1

```python
import requests
from bs4 import BeautifulSoup

url = "https://news.naver.com/"
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'html.parser')

links = soup.find_all('a', class_='cnf_news')
titles = [link.get_text().strip() for link in links]
for title in titles:
    print(title)
```

# 문제 \# 2

```python
import requests
from bs4 import BeautifulSoup
import urllib.request
import re

url = "https://finance.naver.com/sise/"
resp = requests.get(url)
soup = BeautifulSoup(resp.content, 'html.parser')

popular_search_list = soup.find('ul', {'class': 'lst_pop', 'id': 'popularItemList'})
popular_search_items = popular_search_list.find_all('li')

titles = {}
links = {}
cnt = 0

for item in popular_search_items:
    cnt += 1
    link = item.find('a')
    if link:
        titles[cnt] = link.get_text().strip()
        links[cnt] = 'https://finance.naver.com/' + link.get('href')
for number, title in titles.items():
    print(f"{number}: {title}")
```

```python
num = int(input())
sel_url = links[num]
sel_html = urllib.request.urlopen(sel_url)
sel_html_contents = str(sel_html.read().decode("ms949"))

stock_results = re.findall(r"(<dl class=\"blind\">)([\s\S]+?)(</dl>)", sel_html_contents)

sel_stock = stock_results[0]
sel_index = sel_stock[1]

index_list = re.findall(r"(<dd>)([\s\S]+?)(</dd>)", sel_index)

for index in index_list:
    print(index[1])
```
